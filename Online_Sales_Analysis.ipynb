```python
import pandas as pd
df_ol_sales = pd.read_csv('/content/online_sales_dataset.csv')

df_ol_sales
```

```python
#Sales_Analysis
df_ol_sales['TotalSales'] = df_ol_sales['UnitPrice'] * df_ol_sales['Quantity'] * (1 - df_ol_sales['Discount'])
category_sales = df_ol_sales.groupby('Category')['TotalSales'].sum()
print(category_sales)
```

```python
#Customer_Analysis
customer_sales = df_ol_sales.groupby('CustomerID')['TotalSales'].sum()
print(customer_sales)
```

```python
#Time_of_Sale
df_ol_sales['InvoiceDate'] = pd.to_datetime(df_ol_sales['InvoiceDate'])
monthly_sales = df_ol_sales.resample('M', on='InvoiceDate')['TotalSales'].sum()
print(monthly_sales)
```

```python
#Geographical_Analysis
country_sales = df_ol_sales.groupby('Country')['TotalSales'].sum()
print(country_sales)
```

```python
#Product_Analysis
top_products = df_ol_sales.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
#Least_sold_product
least_sold_product = df_ol_sales.groupby('Description')['Quantity'].sum().idxmin()
print(least_sold_product)
```

```python
#Most_sold_product
most_sold_product = df_ol_sales.groupby('Description')['Quantity'].sum().idxmax()
print(most_sold_product)
```

```python
#Return_Analysis
return_rate = df_ol_sales[df_ol_sales['ReturnStatus'] == 'Returned'].shape[0] / df_ol_sales.shape[0]
print(return_rate)
```

```python
#Profit
df_ol_sales['Profit'] = df_ol_sales['TotalSales'] - df_ol_sales['ShippingCost']
print(df_ol_sales[['Description', 'Profit']])
```

```python
# Group by 'Description' and aggregate the necessary columns
grouped = df_ol_sales.groupby('Description').agg({ 'TotalSales': 'sum', 'ShippingCost': 'sum' }).reset_index()

# Calculate profit for each product
grouped['Profit'] = grouped['TotalSales'] - grouped['ShippingCost']

# Display the results
print(grouped)
```

```python
#Category Sales Insights
category_sales = df_ol_sales.groupby('Category')['TotalSales'].sum()
print(category_sales)
```

```python
#Payment_method_Analysis
payment_method_sales = df_ol_sales.groupby('PaymentMethod')['TotalSales'].sum()
print(payment_method_sales)
```
